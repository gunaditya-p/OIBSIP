```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette('husl')
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
#importing and checking data
dataset = pd.read_csv("Iris.csv")
print(dataset.head())
print("\n")
print(dataset.info())
#modifying data
vals = dataset.drop(['Id', 'Species'], axis = 1)
spies = dataset['Species']
vals_train, vals_test, spies_train, spies_test = train_test_split(vals, spies, test_size = 0.20, random_state = 12)
```

```python
sns.violinplot(x='Species', y='SepalLengthCm', data=dataset, inner="quartile")
plt.show()
```

```python
sns.violinplot(x='Species', y='SepalWidthCm', data=dataset, inner="quartile")
plt.show()
```

```python
sns.violinplot(x='Species', y='PetalLengthCm', data=dataset, inner="boxes")
plt.show()
```

```python
sns.violinplot(x='Species', y='PetalWidthCm', data=dataset, inner="boxes")
plt.show()
```

```python
sns.pairplot(dataset.drop(['Id'], axis = 1), hue= 'Species')
```

```python
#Logistic Regression
model1 = LogisticRegression(random_state=0, multi_class='multinomial', solver='newton-cg').fit(vals_train, spies_train)
prediction = model1.predict(vals_test)
print("LOGISTIC REGRESSION: \n")
print(f'Test Accuracy: {accuracy_score(spies_test, prediction)} \n')
print(f'Classification Report: \n {classification_report(spies_test, prediction)}')
```

```python
#Linear Discriminant Analysis
model2 = LinearDiscriminantAnalysis().fit(vals_train, spies_train)
prediction = model2.predict(vals_test)
print("LINEAR DISCRIMINANT ANALYSIS: \n")
print(f'Test Accuracy: {accuracy_score(spies_test, prediction)} \n')
print(f'Classification Report: \n {classification_report(spies_test, prediction)}')
```

```python
#K Nearest Neighbours
model3 = KNeighborsClassifier().fit(vals_train, spies_train)
prediction = model3.predict(vals_test)
print("K NEAREST NEIGHBOURS: \n")
print(f'Test Accuracy: {accuracy_score(spies_test, prediction)} \n')
print(f'Classification Report: \n {classification_report(spies_test, prediction)}')
```

```python
#Descision Tree
model4 = DecisionTreeClassifier().fit(vals_train, spies_train)
prediction = model4.predict(vals_test)
print("DESCISION TREE: \n")
print(f'Test Accuracy: {accuracy_score(spies_test, prediction)} \n')
print(f'Classification Report: \n {classification_report(spies_test, prediction)}')
```

```python
#Gaussian Naive Bayes
model5 = GaussianNB().fit(vals_train, spies_train)
prediction = model5.predict(vals_test)
print("GAUSSIAN NAIVE BAYES: \n")
print(f'Test Accuracy: {accuracy_score(spies_test, prediction)} \n')
print(f'Classification Report: \n {classification_report(spies_test, prediction)}')
```

```python
#Support Vector Machine
model6 = SVC().fit(vals_train, spies_train)
prediction = model6.predict(vals_test)
print("SUPPORT VECTOR MACHINE: \n")
print(f'Test Accuracy: {accuracy_score(spies_test, prediction)} \n')
print(f'Classification Report: \n {classification_report(spies_test, prediction)}')
```
